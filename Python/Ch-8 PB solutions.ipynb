```python
#537
# Write a python program that has a class Point with attributes as the x and y co-ordinates.
# 1. Add a method ‘distance from origin’ to class Point which returns the distance of the given point from origin. The
# equation is
# 2. Add a method ‘translate’ to class Point, which returns a new posi􀆟on of point a􀅌er transla􀆟on
# 3. Add a method ‘reflect_x’ to class Point, which returns a new point which is the reflec􀆟on of the point about the x-axis.
# 4. Add a method ‘distance’ to return distance of the given point with respect to the other point. The formula for calcula􀆟ng
# distance between A(x1,y1) and B(x2,y2) is
class Point:
    def __init__(self,p):
        self.p=p

    def distance_origin(self):
        d=(self.p[0]**2+self.p[1]**2)**0.5
        return d
    
    def translate(self,t):
        q=(self.p[0]+t[0],self.p[1]+t[1])
        return q 

    def reflect_x(self):
        q=(self.p[0],self.p[1]*(-1))
        return q

    def distance_point(self,q):
        d=((self.p[0]-q[0])**2+(self.p[1]-q[1])**2)**0.5
        return d

point=Point((1,2))
print(point.distance_origin())
print(point.translate((1,1)))
print(point.reflect_x())
point1=Point((2,-3))
print(point1.distance_point((3,4)))
```
