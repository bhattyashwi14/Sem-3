```python
# Q637 Write a program that overload the + operator so that it can add two object of class fraction
# Q638 Write a program that overload the * operator so that it can add two object of class fraction
class Fraction:
    def __init__(self,x,y):
        self.num=x;
        self.dino=y;
    def __add__(self,other):
        a=self.num*other.dino+other.num*self.dino;
        b=self.dino*other.dino;
        return f"{a}/{b}"
    def __mul__(self,other):
        a=self.num*other.num;
        b=self.dino*other.dino
        return f"{a}/{b}"
    def __str__(self):
        return f"{self.num}/{self.dino}"

f1=Fraction(1,3);
f2=Fraction(1,2);
print(f1);
print(f2);
print()
print(f1+f2)
print(f1*f2)
```

```python
# Q639 Write a program to find the distance between two points in cartesian cordinate system
class Distance:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def distance(self,other):
        X=self.x-other.x
        Y=self.y-other.y
        dist=(X**2+Y**2)**0.5
        return dist
p1=Distance(1,3)
p2=Distance(3,1)
print(p1.distance(p2))
```

```python
# Q641 Create a class student with following member attributes: roll no, name, age and total marks. Create suitable methods for
# reading and printing member variables. Write a python program to overload ‘==’ operator to print the details of students
# having same marks.
class student:
    def __init__(self,r,n,a,m):
        self.m=m
        self.r=r
        self.n=n
        self.a=a;
    def display(self, other):
        print(f"Name={self.n},RollNo:{self.r},Age:{self.a},Marks:{self.m}")
        print(f"Name={other.n},RollNo:{other.r},Age:{other.a},Marks:{other.m}")
    def __eq__(self,other):
        if (self.m==other.m):
            self.display(other)
            return True
        else:
            print("Not same marks")
            return False
        

s1=student(2,"yashi",19,95)
s2=student(18,"vish",19,95)
s1==s2
```

```python
# Q642 Write a program to create a class called Data having “value” as its data member. Overload the (>) and the (<) operator for
# the class. Instantiate the class and compare the objects using _lt_ and _gt_.
class Data:
    def __init__(self,value):
        self.value=value
    def __gt__(self,other):
        if self.value>other.value:
            print(f"{self.value} wins gt")
        else:
             print(f"{other.value} wins gt")

        return self.value>other.value
    def __lt__(self,other):
        if self.value<other.value:
            print(f"{self.value} wins lt")
        else:
             print(f"{other.value} wins lt")
        return self.value<other.value

d1=Data(7)
d2=Data(9)
print(d1>d2)
print(d1<d2)
        
```
