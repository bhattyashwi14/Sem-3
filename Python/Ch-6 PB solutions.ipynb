```python
#475
#Write a “pager” program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing
# each time to ask the user to enter the word “continue”, in order to show the next 25 lines or enter the word “stop” to
# close the file.
f=open("Pb475.txt")
L=f.readlines()
c=0
for i in range(0,min(25,len(L))):
    print(L[i],end="")
    c=i
while True:
    ans=input("Do you want to read further?")
    if ans.lower()=="continue":
        for i in range(c+1,c+26):
            if i<len(L):
                print(L[i],end="")
                c=i
            else:
                print("Reached to the end of the file!")
                break
    elif ans.lower()=="stop":
        break
f.close()
```

```python
#478
# Write a python program to read a text file “Story.txt” and print only word starting with ‘I’ in reverse order.
# Example: If value in text file is : ‘INDIA IS MY COUNTRY’
# Output will be: ‘AIDNI SI MY COUNTRY’
f=open("story.txt")
L=f.readlines()
st=""
for i in range(len(L)):
    l=L[i].split()
    for j in range(len(l)):
        if l[j][0]=='I':
            st+=l[j][::-1]+" "
        else:
            st+=l[j]+" "
print(st)
f.close()
```

```python
#479
#Write a Python program to count words, characters and spaces from text file.
f=open("pb479.txt")
L=f.readlines()
word=0
char=0
sp=0
for i in range(len(L)):
    l=L[i]
    y=L[i].split()
    word+=len(y)
    char+=len(L[i])-1
    for j in range(len(l)):
        if l[j]==" ":
            sp+=1
print("words:",word)
print("characters:",char)
print("spaces:",sp) 
f.close()
```

```python
#483
# Write a python program to read through the mbox-short.txt and figure out who has
# sent the greatest number of mail messages. The program looks for 'From ' lines and
# takes the second word of those lines as the person who sent the mail. The program
# creates a Python dictionary that maps the sender's mail address to a count of the
# number of times they appear in the file. After the dictionary is produced, the program reads through the dictionary to
# identify the sender with the maximum count (the most prolific sender).
f=open("mbox-short.txt")
d={}
L=f.readlines()
for i in range(len(L)):
    l=L[i].split()
    if len(l)>1 and l[0]=="From":
        d[l[1]]=d.get(l[1],0)+1
print(d)
y=sorted(d.items(),key=lambda x:x[1])[-1]
print()
print("Highest mails sent:",y)
```
